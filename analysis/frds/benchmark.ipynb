# Benchmarking FRDS

Theoretically, an FID of 0 indicates that the generated images are of "perfect" or "ideal" quality. In practice, this only happens if the generated images are identical to the real images.

The same principle applies to FRDS. In this notebook, we compute the FRDS between the train set and test set. Here, the test set acts as the "ideal" synthetic data. We investigate what the ideal FRDS is.
